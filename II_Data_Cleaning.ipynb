```python
# First we import the python software libraries that we will use here
# Import Pandas library
import pandas as pd

# Import numpy library
import numpy as np
```

```python
#importing the data set into the terminal

demographic = pd.read_csv('/content/demographic.csv')
demographic.head(5)
```

# Validity

```python
# Checking for missing values in the dataset demographic data.

# Creating a function to check
def missing_values(data):

  # identifying the missing values from the demographic data dataset
  #
  miss = data.isnull().sum().sort_values(ascending = False)
  # finding the percentage of missing values
  percentage = (data.isnull().sum() / len(data)).sort_values(ascending = False)
  # storing  total missing values in percentage in dataframe
  #
  missing_data = pd.DataFrame({'Total' : miss, 'Percentage' : percentage})
  #if the percentage is 0, indicates no missing values hence removed
  #
  missing_data.drop(missing_data[missing_data['Percentage'] == 0].index, inplace = True)

  return missing_data

missing_values(demographic)
```

```python
demographic = demographic.dropna()
```

```python
# Checking for outliers
# checking to see if any values appear True if so they are outliers
#
Q1 = demographic.quantile(0.25)
Q3 = demographic.quantile(0.75)

IQR = Q3 - Q1


(demographic < (Q1 - 1.5 * IQR)) |(demographic > (Q3 + 1.5 * IQR))
```

```python
# Sorting the data so that we have them in ascending order based on the date of birth

demographic = demographic.sort_values(
     by="DOB",
     ascending=True
 )

demographic
```

```python
# Dropping the iregular value in the data set
demographic = demographic.drop(axis = 0, labels = [1573])
```

```python
# Removal of irrelevant data

# Creating a function to drop unnecessary columns
# The dataframe was named df_demographic
def drop_columns(data, cols):
  #drop columns not required
  data.drop(cols, axis = 1, inplace = True)

  return data

#columns to be dropped
columns = ['Unnamed: 0']
demographic = drop_columns(demographic, columns)
```

```python
# Changing the data type of first_name, last_name, gender, job_title, job_title, 
# job_industry_category, wealth_segment
# to string values

demographic['first_name'] = demographic['first_name'].astype('string', '|S')
demographic['last_name'] = demographic['last_name'].astype('string', '|S')
demographic['gender'] = demographic['gender'].astype('string', '|S')
demographic['job_title'] = demographic['job_title'].astype('string', '|S')
demographic['job_industry_category'] = demographic['job_industry_category'].astype('string', '|S')
demographic['wealth_segment'] = demographic['wealth_segment'].astype('string', '|S')
```
